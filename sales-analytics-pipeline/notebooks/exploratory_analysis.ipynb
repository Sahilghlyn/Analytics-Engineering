```python
# Exploratory Analysis for Sales Analytics Pipeline
import pandas as pd
import matplotlib.pyplot as plt

# Load aggregated data
monthly = pd.read_csv("../data/monthly_sales.csv")
top_products = pd.read_csv("../data/top_products.csv")

# Total sales per month
plt.figure(figsize=(8,5))
plt.plot(monthly['order_date'], monthly['total_sales'], marker='o')
plt.title('Total Sales Per Month')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

# Top 5 products by revenue
plt.figure(figsize=(8,5))
plt.bar(top_products['product'], top_products['total_sales'], color='orange')
plt.title('Top 5 Products by Revenue')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.show()
```
